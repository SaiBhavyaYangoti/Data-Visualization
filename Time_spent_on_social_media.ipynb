```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data=pd.read_csv("/content/sample_data/dummy_data.csv")
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull()
```

Platform

```python
d=data.groupby(by=['platform']).agg({'time_spent':'mean'}).reset_index()
fig=px.bar(d,x='platform',y='time_spent',color='time_spent',title='Average Time spent on Social media platforms')
fig.update_layout(width=500, height=500)
fig.show()
```

```python
d=data.groupby(by=['platform']).agg({'time_spent':'mean'}).reset_index()
fig = px.pie(d, values='time_spent', names='platform', title='Average Time spent on Social media platforms')
fig.update_layout(width=500, height=400)
fig.show()
```

Gender

```python
#bar
d=data.groupby(by=['gender']).agg({'time_spent':'mean'}).reset_index()
fig=px.bar(d,x='gender',y='time_spent',color='time_spent',title='Time Spent on Social Media by Gender')
fig.update_layout(width=500, height=500)
fig.show()
```

```python
#pie
d=data.groupby(by=['gender']).agg({'time_spent':'mean'}).reset_index()
fig = px.pie(d, values='time_spent', names='gender', title='Time Spent on Social Media by Gender')
fig.update_layout(width=400, height=400)
fig.show()
```

```python
#group bar
d=data.groupby(by=['gender','platform']).agg({'time_spent':'mean'}).reset_index()
fig=px.bar(d,x='gender',y='time_spent',color='platform',barmode='group',title='Average Time Spent on Social Media Platforms by gender')
fig.update_layout(width=700, height=500)
fig.show()
```
